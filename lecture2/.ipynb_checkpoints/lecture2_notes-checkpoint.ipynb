# Lecture 2: Linear Regression

First we load in packages and data

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing
from sklearn.preprocessing import PolynomialFeatures

data = pd.read_csv('data/country-stats.csv')
data.head()
```

## Supervised Learning

In Supervised Learning, we have:
 - Input Features $x_1,x_2,x_3,...,x_n$, where each is a set of values
 - Known (desired) outputs $y_1,y_2,y_3,...,y_n$
 - A Dataset of $<x_i,y_i>$ pairs
 
The goal is the learn the mapping $f \colon X \rightarrow Y$, such that $y_i = f(x_i)$ for all $i = 1,2,3,...,n$

These problems come in two types:

 - Continuous Problems
      - Here, the desired labels are **continuous** - i.e. they can take any **value** from some range.
      - In this case, we use **Regression**.
      - Examples include predicting company stock price given company earnings and revenue, or predicting the price of a house given its size and age
 - Discrete Problems
      - Here, the desire labels are **discrete** - i.e. they take from a **discrete set**.
      - In this case, we use **Classification**.
      - Examples include predicting a digit label given an image of a handwritten digit, or predicting sentiment given a text.

## Finding a Good Model

The simplest model for $f \colon X \rightarrow Y$ is $y=x$, which can actually work very well in some cases.

For example, if both $X$ and $Y$ are strongly correlated and fit inside the same range, it works well. For the dataset, this would be plotting **Estimated Corruption vs Government Effectiveness** as below.

Here we simply plot a scatter plot of the data using plt.scatter, add titles and then add a $y=x$ line:

```python
# Data
plt.scatter(data["Estimated Control of Corruption (scale -2.5 to 2.5)"], 
            data["Estimated Government Effectiveness (scale -2.5 to 2.5)"])

# y = x line
x_sample = np.linspace(-2.5, 2.5, 100)
y = x_sample
plt.plot(x_sample, y, color='orange')

# Add titles
plt.title('Estimated Corruption vs Government Effectiveness')
plt.xlabel("Estimated Control of Corruption")
plt.ylabel("Estimated Government Effectiveness")

# Show in notebook
plt.show()
```

However for most practical applications this is not true, shown below.

Here we simply instead plot **Estimated Government Corruption vs Urban Population**, two features you'd expect to have no real correlation nor the same scale:

```python
# Data
plt.scatter(data["Estimated Control of Corruption (scale -2.5 to 2.5)"], 
            data["Urban Population (%)"])

# y = x
x_sample = np.linspace(-2.5, 2.5, 100)
y = x_sample
plt.plot(x_sample, y, color='orange')

# Titles
plt.title('Estimated Corruption vs Urban Population')
plt.xlabel("Estimated Control of Corruption")
plt.ylabel("Urban Population (%)")

# Show in notebook
plt.show()
```

We could also try a more flexible model: $y = ax$

Here we build a very simple Linear Regresiion model and allow it to adjust the $a$ value to match the general **slope** of the data:

**Notice how we set fit_intercept=False, stopping the model for realigning predictions based on the y-intercept**

```python
# Data
plt.scatter(data["Estimated Control of Corruption (scale -2.5 to 2.5)"], 
            data["Urban Population (%)"])

# Now need to create a y = ax model
model = LinearRegression(fit_intercept=False)


# Set columns as 
X = data["Estimated Control of Corruption (scale -2.5 to 2.5)"].values.reshape(-1,1)
Y = data["Urban Population (%)"]
model.fit(X, Y)

x_sample = np.linspace(-2.5, 2.5, 100)
plt.plot(x_sample, model.predict(x_sample.reshape(-1,1)), color='orange')

# Finishing touches
plt.title('Estimated Corruption vs Urban Population')
plt.xlabel("Estimated Control of Corruption")
plt.ylabel("Urban Population (%)")
plt.show()
```

This graph gave us a correct looking general shape, but it's in the wrong place and would be a terrible predictor

We could instead try $y = b$, allowing flexibility on the **position** of the line but not the slope:

```python
plt.scatter(data["Estimated Control of Corruption (scale -2.5 to 2.5)"], 
            data["Urban Population (%)"])
plt.title('Estimated Corruption vs Urban Population')
plt.xlabel("Estimated Control of Corruption")
plt.ylabel("Urban Population (%)")

model = LinearRegression(fit_intercept=True)
X = np.zeros((len(data["Urban Population (%)"]), 1), dtype=float)
Y = data["Urban Population (%)"]
model.fit(X, Y)

x_sample = np.linspace(-2.5, 2.5, 100)
plt.plot(x_sample, model.predict(x_sample.reshape(-1,1)), color='orange')
plt.show()
```

However again this would be a useless predictor. While being around the data points, it does not follow the pattern of the data.

This leads us to the simplest useful model: $y = ax + b$. This allows the line to be shaped both to the **slope** of the data and the **position** as below (done by setting fit_intercept=True):

```python
plt.scatter(data["Estimated Control of Corruption (scale -2.5 to 2.5)"], 
            data["Urban Population (%)"])
plt.title('Estimated Corruption vs Urban Population')
plt.xlabel("Estimated Control of Corruption")
plt.ylabel("Urban Population (%)")

model = LinearRegression(fit_intercept=True)
X = data["Estimated Control of Corruption (scale -2.5 to 2.5)"].values.reshape(-1,1)
Y = data["Urban Population (%)"]
model.fit(X, Y)

x_sample = np.linspace(-2.5, 2.5, 100)
plt.plot(x_sample, model.predict(x_sample.reshape(-1,1)), color='orange')
plt.show()
```

## Optimizing Linear Regression: Gradient Descent

Gradient Descent is about using the gradient of the data to slowly improve the RSME - The **Root Mean Square Error**

For each $y_i$, we have a predicted value $p_i$, which is generated with the Linear Regression model $y = ax + b$. Therefore for each individual error, we have a **error** of $p_i-y_i$. **By squaring this value, we get something that punishes large errors stronger, and makes small errors even less significant**

So for a set of predictions, we can define the following equation:
\begin{equation*}
E = \frac{1}{2} \sum_{i=1}^{M} (p_i - y_i)^2
\end{equation*}

This means that a smaller E means our predictions where closer to the real values.

Then the $RSME = \sqrt{\frac{E}{M}}$, which becomes a sort of "average error per prediction", but as above penalising larger errors more.

Now clearly, the RSME is something we'd want to **minimise**. Gradient Descent allows us to do this.

Gradient descent works by changing a and b in the function $y = ax + b$, based on the data and the RSME (the loss function). We repeatedly update parameters a and b by taking small steps in the direction of the partial derivative (change in a with respect to E and same for b).

So at each stage, $a := a - \alpha \frac{\delta E}{\delta a}$ is applied where $\alpha = $ learning rate / step size.
Similar for b.

The code for this is shown below, for **GDP vs Enrolment Rate** (prediciting GDP given Enrolment Rate):

```python
# Define a function for plotting a Linear Regression model given features and outputs.
def plot_simple_linear_regression(X, Y, a, b, title, xlabel, ylabel):
    plt.scatter(X, Y)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    x_sample = np.linspace(np.min(X), np.max(X), 100)
    plt.plot(x_sample, x_sample*a + b, color='orange')
    plt.show()
```

```python
# Define a function applying Gradient Descent to inputs X, outputs Y, linear model coefficients a and b, and a learning rate
def gradient_descent_long(X,Y,a,b,learning_rate):
    for epoch in range(10):
        update_a = 0.0
        update_b = 0.0
        error = 0.0
        for i in range(len(Y)):
            y_predicted = a * X[i] + b
            update_a += (y_predicted - Y[i])*X[i]
            update_b += (y_predicted - Y[i])
            error += np.square(y_predicted - Y[i])
        a = a - learning_rate * update_a
        b = b - learning_rate * update_b
        rmse = np.sqrt(error / len(Y))
        print("a: {:.4f}\tb: {:.9f}\tRMSE: {:.4f}".format(a, b, rmse))
    return a,b
```

```python
X = data["GDP per Capita (PPP USD)"].values
Y = data["Enrolment Rate, Tertiary (%)"].values

a = 0.0
b = 0.0
learning_rate = 1e-11
a,b = gradient_descent_long(X,Y,a,b,learning_rate)
plot_simple_linear_regression(X, Y, a, b, 
                              "GDP vs Enrolment Rate", 
                              "GDP per Capita (PPP USD)", 
                              "Enrolment Rate, Tertiary (%)")
```

### The Gradient

It can often be more convenient to work with vector notation. 

We then define the gradient as the **vector of all partial derivatives**.

I.e. for a function of n reals (the features) to one real (the output), the gradient of f is defined as an n x 1 vector, where each element xi is the partial derivative of f with respect to the i-th feature.

Therefore, we can also create a more compact version, taking advantage of python vector features:

```python
def gradient_descent_compact(X,Y,a,b,learning_rate):
    for epoch in range(10):
        y_predicted = a * X + b
        a = a - learning_rate * ((y_predicted - Y)*X).sum()
        b = b - learning_rate * (y_predicted - Y).sum()
        rmse = np.sqrt(np.square(y_predicted - Y).mean())
        print("RMSE: " + str(rmse))
```

## Optimizing Linear Regression: The Analytical Solution with Scikit-Learn

Using vectors leads us to an analytical method. **Note however that this solution requires matrix inversion, and is therefore not so great for large problems since this has cubic complexity**.

$X = \begin{bmatrix}
    x_1 & 1.0 \\
    x_2 & 1.0 \\
    \vdots & \vdots \\
    x_M & 1.0
    \end{bmatrix}$
$y = \begin{bmatrix}
    y_1 \\
    y_2 \\
    \vdots \\
    y_M
    \end{bmatrix}$
$\theta = \begin{bmatrix}
    a \\
    b
    \end{bmatrix}$

$\nabla_\theta E(\theta) = X^T (X\theta - y) = 0$

Which gives us

$\Longrightarrow \theta^* = (X^T X)^{-1} X^T y$

This is great for directly finding the optimal parameter values, but as above it is not great for large problems.

We can use this to instead calculate the analytical least-squares solution using Scikit-Learn, as below:

```python
# Imports
from sklearn.linear_model import LinearRegression

# Fit model to data
model = LinearRegression(fit_intercept=True)
X = data["GDP per Capita (PPP USD)"].values.reshape(-1,1)
Y = data["Enrolment Rate, Tertiary (%)"]
model.fit(X, Y)

# Get Mean Squared Error using vectors
mse = np.square(Y - model.predict(X)).mean()
print("RMSE: " + str(np.sqrt(mse)))

# Plot data
plt.scatter(data["GDP per Capita (PPP USD)"], 
            data["Enrolment Rate, Tertiary (%)"])

# Plot Regression Line
x_sample = np.linspace(0, 80000, 100)
plt.plot(x_sample, 
         model.predict(x_sample.reshape(-1,1)), 
         color='orange')

# Add titles and show plot
plt.title('GDP vs Enrolment Rate')
plt.xlabel("GDP per Capita (PPP USD)")
plt.ylabel("Enrolment Rate, Tertiary (%)")
plt.show()
```

## Multiple Linear Regression

So far we have only looked at predicting the output using one input feature. How about if we want to use multiple features?

Well it turns out we basically always want to use multiple features, since they may all effect the output in various ways. Using them all gives us a better chance of finding patterns and correlations, and ultimately predicting correctly.

So here we will plot **GDP vs Enrolment Rate** but we're actually using every feature expect the output feature itself, and any text columns that don't help:

```python
# Fit model to data
model = LinearRegression(fit_intercept=True)
# Get all columns to be used as features
X = data.copy().drop(["Country Name", "Enrolment Rate, Tertiary (%)"], axis=1)
Y = data["Enrolment Rate, Tertiary (%)"]
model.fit(X, Y)

# Get Mean Squared Error
mse = np.square(Y - model.predict(X)).mean()
print("RMSE: " + str(np.sqrt(mse)))

# Plot Data
plt.scatter(data["GDP per Capita (PPP USD)"], 
            data["Enrolment Rate, Tertiary (%)"])

# Plot actual GDP vs predicted Enrolment Rate
x_pred = data["GDP per Capita (PPP USD)"]
plt.plot(x_pred, model.predict(X), '.', color='orange')

# Add titles and show plots
plt.title('GDP vs Enrolment Rate')
plt.xlabel("GDP per Capita (PPP USD)")
plt.ylabel("Enrolment Rate, Tertiary (%)")
plt.show()
```

## Exploring the Parameters

Now we have a model that seems to predict reasonably well. What we can do now is actually investigate the parameters used to predict the output in more detail, to see which features were the most useful/ correlated to the output.

We might do this by showing the **optimized coefficients for each of the input features** as below with a few lines of code once the model is built:

```python
headers=list(X)
coefficients = []
for i in range(len(headers)):
    coefficients.append({"Property": headers[i], 
                         "coefficient": model.coef_[i]})
pd.DataFrame(coefficients)
```

However, there's a problem with this. Namely, each feature is in a different range, and therefore a multiplicative coefficient is going to be different based on that as well as how predictive the feature is.

Therefore we need to **scale each input feature before model fitting**, which will make the coefficients more comparable.

One way to scale inputs is to map them to a range of -1 to 1, where values around 0 are close to the mean value for that columns, and values closer to -1 or 1 are on the outer ranges of the data. This can be done with **sklearn.preprocessing.scale**. Here's an example of how we'd do this just for the GDP column:

```python
# Get Data
Z = pd.DataFrame(data, columns=["GDP per Capita (PPP USD)"])

# Scale Data
Z_scaled = preprocessing.scale(Z)

# Put output in a nice format and print head
asd = pd.concat([Z, pd.DataFrame(Z_scaled)], axis=1)
asd.columns = ["Z", "Z_scaled"]
asd.head()
```

Now we can apply this method to every feature column **before model fitting**, as below:

```python
# Create model
model = LinearRegression(fit_intercept=True)

# Get feature columns
X = data.copy().drop(["Country Name", "Enrolment Rate, Tertiary (%)"], axis=1)

# Scale ALL feature columns
X_scaled = preprocessing.scale(X)

# Get output column, and fit model
Y = data["Enrolment Rate, Tertiary (%)"]
model.fit(X_scaled, Y)

# Get Mean Squared Error
mse = np.square(Y - model.predict(X_scaled)).mean()
print("RMSE: " + str(np.sqrt(mse)))

# Plot Data
plt.scatter(data["GDP per Capita (PPP USD)"], 
            data["Enrolment Rate, Tertiary (%)"])

# Plot predictions
x_pred = data["GDP per Capita (PPP USD)"]
plt.plot(x_pred, model.predict(X_scaled), '.', color='orange')

# Add labels and show plot
plt.title('GDP vs Enrolment Rate')
plt.xlabel("GDP per Capita (PPP USD)")
plt.ylabel("Enrolment Rate, Tertiary (%)")
plt.show()
```

Now we can look at the coefficients and actually compare them:

```python
headers=list(X)
coefficients = []
for i in range(len(headers)):
    coefficients.append({"Property": headers[i], "coefficient": model.coef_[i]})
pd.DataFrame(coefficients)
```

The coefficients represent how much that feature affects the output. Therefore the coefficients furthest away from zero represent the most useful predictors of the output.

Here we can see that Population Growth Rate, Urban Population and Fertility Rate are all much better predictors of Enrolment Rate, and features such as Unemployment, Infant Mortality and Population Density are not very useful predictors.

## Polynomial Regression

We can take our existing input features and have them transformed into a series of polynomial features. This allows us to look at the interactions between sets of features, and how good they are at predicting as a set.

For example if we work with **degree 2** polynomials, we will be able to capture **pairwise feature interactions**.

With degree 2, the features $[z_1,z_2]$ becomes $[1,z_1,z_2,z_1^2,z_1 z_2,z_2^2]$

And with the 11 features in our country stats dataset, we get 78 polynomial features that capature all the pairwise feature interactions.

Polynomial regression is a special case of multiple linear regression - as a statistical estimation problem it is linear, but because it uses polynomial features it is able to model non-linear data.

Using Polynomial Regression is as simple as using **sklearn.preprocessing.PolynomialFeatures** to create our superset of features, and then feeding them into the same LinearRegression model as used throughout this code:

```python
# Create model and collect raw data
model = LinearRegression(fit_intercept=True)
X = data.copy().drop(["Country Name", "Enrolment Rate, Tertiary (%)"], axis=1)
Y = data["Enrolment Rate, Tertiary (%)"]

# Create the PolynomialFeatures object, and apply it to the X data
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Fit the model using the polynomial features
model.fit(X_poly, Y)

# Get the Mean Squared Error
mse = np.square(Y - model.predict(X_poly)).mean()
print("RMSE: " + str(np.sqrt(mse)))

# Plot Raw Data
plt.scatter(data["GDP per Capita (PPP USD)"], 
            data["Enrolment Rate, Tertiary (%)"])

# Plot Predictions
x_pred = data["GDP per Capita (PPP USD)"]
plt.plot(x_pred, model.predict(X_poly), '.', color='orange')

# Add labels and show plot
plt.title('GDP vs Enrolment Rate')
plt.xlabel("GDP per Capita (PPP USD)")
plt.ylabel("Enrolment Rate, Tertiary (%)")
plt.show()
```

As you can see, this is the best (lowest) RSME we have achieved so far at 13.6692.

However, this can be a dangerous metric to look at solely, as we'll discuss below.

## Overfitting

Say we now try generating and using 3rd degree polynomial features, producing 364 input polynomial features from the original 11:

```python
# Create model using polynomial features
model = LinearRegression(fit_intercept=True)
X = data.copy().drop(["Country Name", "Enrolment Rate, Tertiary (%)"], axis=1)
poly = PolynomialFeatures(degree=3)
X_poly = poly.fit_transform(X)
Y = data["Enrolment Rate, Tertiary (%)"]
model.fit(X_poly, Y)

# Get Mean Squared Error
mse = np.square(Y - model.predict(X_poly)).mean()
print("RMSE: " + str(np.sqrt(mse)))

# Plot raw Data
plt.scatter(data["GDP per Capita (PPP USD)"], 
            data["Enrolment Rate, Tertiary (%)"])

# Plot Predictions
x_pred = data["GDP per Capita (PPP USD)"]
plt.plot(x_pred, model.predict(X_poly), '.', color='orange')

# Add Labels and show plot
plt.title('GDP vs Enrolment Rate')
plt.xlabel("GDP per Capita (PPP USD)")
plt.ylabel("Enrolment Rate, Tertiary (%)")
plt.show()
```

Wow, see how perfectly that predicted each point! 0.00018 RSME! Yeah, there's a problem here. We are predicting the outputs **too well**. This means that we're relying far too much on the input data, **overfitting** to it, meaning our model wouldn't be great at predicting new situations outside this dataset, which is ultimately we want it to be good at.

There are twice as many features for each datapoint as there are datapoints in the whole dataset. Such a model often has the capacity to **memorize the training data**, but that doesn't mean it is able to properly capture the **underlying patterns**.

We can demonstrate this issue further by splitting the data:

## Dataset Splits for Model Selection

We want to split our dataset into 3 sets: A **Training Set** used for training your models and fitting the parameters, a **Development Set** for continuous evaluation and hyperparameter selection, and a **Test Set** for realistic evaluation once the training and tuning is done:

```python
from sklearn.model_selection import train_test_split
# Seperate out 20% of raw data for testing
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, train_size=0.8, random_state=1)
# Seperate out 20% of remaining data for development
X_train, X_dev, y_train, y_dev = train_test_split(X_train, y_train, test_size=0.2, train_size=0.8, random_state=1)

print("Total Shape: {}".format(X.shape))
print("Training Set: {}".format(X_train.shape))
print("Development Set: {}".format(X_dev.shape))
print("Test Set: {}".format(X_test.shape))
```

We can define a function for fitting polynomial regression with different degrees:

```python
def fit_polynomial_regression(X, Y, degree):
    model = LinearRegression(fit_intercept=True)
    poly = PolynomialFeatures(degree=degree)
    X_poly = poly.fit_transform(X)
    model.fit(X_poly, Y)
    return model
```

Let's define one more function, to help us evaluate and plot the output from trained polynomial models:

```python
def plot_gdp_vs_enrolment_rate(X, Y, degree, model):
    plt.scatter(X["GDP per Capita (PPP USD)"], Y)
    plt.title('GDP vs Enrolment Rate')
    plt.xlabel("GDP per Capita (PPP USD)")
    plt.ylabel("Enrolment Rate, Tertiary (%)")
    
    poly = PolynomialFeatures(degree=degree)
    X_poly = poly.fit_transform(X)
    
    mse = np.square(Y - model.predict(X_poly)).mean()
    print("RMSE: " + str(np.sqrt(mse)))
    
    x_pred = X["GDP per Capita (PPP USD)"]
    plt.plot(x_pred, model.predict(X_poly), '.', color='orange')

    plt.show()
```

Now we can very quickly demonstrate the problem with overfitting. First we'll train the model using 3rd degree polynomial regression, and then we'll see how well it performs on the development set, plotting GDP vs Enrolment Rate along the way:

```python
model = fit_polynomial_regression(X_train, y_train, degree=3)
plot_gdp_vs_enrolment_rate(X_train, y_train, degree=3, model=model)
```

```python
model = fit_polynomial_regression(X_train, y_train, degree=3)
plot_gdp_vs_enrolment_rate(X_dev, y_dev, degree=3, model=model)
```

Here we can see the problem. The model was perfectly predicting the values in the training set, almost as if it has memorised the solutions, with a RSME around zero. However once we used that model to predict the values in the **Development Set**, the RSME was useless. All the values were predicted what seems to be at almost random. This is because the model was **overfit** to the training data, and therefore did not react well to new unseen data.

Let's try the same thing but instead with 2nd degree polynomial regression:

```python
model = fit_polynomial_regression(X_train, y_train, degree=2)
plot_gdp_vs_enrolment_rate(X_dev, y_dev, degree=2, model=model)
```

This is better, but some predictions are still wildly off, and the RMSE is very far off what we see from the training set. Let's instead try 1st degree polynomial regression:

```python
model = fit_polynomial_regression(X_train, y_train, degree=1)
plot_gdp_vs_enrolment_rate(X_dev, y_dev, degree=1, model=model)
```

And this is indeed the model that gives the smallest error on held-out data.

If we had a **much larger dataset**, with thousands or tens of thousands of examples, then **the model wouldn't be able to overfit to the training set so easily** and the **polynomial features might actually provide a benefit**.

But since we had so few data points, the higher degree polynomial regression models could just learn the training set.

This is also true if we have far less features - the higher degree polynomial features can help us get the shape of the data better in these cases.

We can try polynomial regression with only one variable again, the GDP:

```python
# Get the raw data
X_single = pd.DataFrame(data, columns=["GDP per Capita (PPP USD)"])
Y_single = data["Enrolment Rate, Tertiary (%)"]

# Create the training, development and testing dataset splits
X_train_single, X_test_single, y_train_single, y_test_single = train_test_split(X_single, Y_single, test_size=0.2, train_size=0.8, random_state=1)
X_train_single, X_dev_single, y_train_single, y_dev_single = train_test_split(X_train_single, y_train_single, test_size=0.2, train_size=0.8, random_state=1)

# Fit the model and plot the training and development prediction plots
model = fit_polynomial_regression(X_train_single.values.reshape(-1,1), y_train_single, degree=3)
plot_gdp_vs_enrolment_rate(X_train_single, y_train_single, degree=3, model=model)
plot_gdp_vs_enrolment_rate(X_dev_single, y_dev_single, degree=3, model=model)
```

Here you can see that the RSME for each are much closer, implying that our model is much closer to understanding the **underlying patterns**. Since if both RSME's were exactly the same, assuming the data isn't causing that and it actually predicts *well* in both cases, it would imply that our model is as good as we can get it - i.e. it predicts new data as well as it predicts data used in training.

## Annotating Graphs

We can annotate specific entries in a plot by calling **plt.annotate** as below:

```python
# Plot raw data
plt.scatter(data["GDP per Capita (PPP USD)"], 
            data["Enrolment Rate, Tertiary (%)"])

# Annotate specific countries
countries = list(data["Country Name"])
for c in ["UK", "USA", "South Korea", "Luxembourg", "Qatar", "Malawi"]:
    country_num = countries.index(c) # get data point index for this country
    plt.annotate(countries[country_num], (data["GDP per Capita (PPP USD)"][country_num], data["Enrolment Rate, Tertiary (%)"][country_num]))

# Add labels and show plot
plt.title('GDP vs Enrolment Rate')
plt.xlabel("GDP per Capita (PPP USD)")
plt.ylabel("Enrolment Rate, Tertiary (%)")
plt.show()
```
